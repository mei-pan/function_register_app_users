```python
# Start coding here
# Use as many cells as you need
def validate_user(name, email, password): 
    if validate_name(name) == False:
        raise ValueError('user name is invalid, must be greater than 2 characters')
    if validate_email(email) == False:
        raise ValueError("email address is invalid")
    if validate_password(password) == False:
        raise ValueError("password is incorrect")
    return True
```

```python
# creating a user checking function that returns user information if all parameters match, otherwise inform user where to check 
def register_user(name, email, password):
    def validate_user(name, email, password): 
        if validate_name(name) == False:
            raise ValueError('user name is invalid')
        if validate_email(email) == False:
            raise ValueError("email address is invalid")
        if validate_password(password) == False:
            raise ValueError("password is invalid, must be at least 8 characters include a captial letter, a number between 0-9")
        return True
    
    try:
        validate_user(name, email, password)
        user_info = {'name': name, 'email': email, 'password': password}
        print('Registeration Success!')
        return user_info
    except ValueError as e: 
        print(f'Error: {e}. Please modify your input')

# Note: The functions validate_name, validate_email, and validate_password need to be defined for this code to work.
```
